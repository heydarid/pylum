```python
class FDEModes:
    def __init__(self,apiObject,width,height,etch,isTE,wavl,trialModes):
        # Open fresh new file, write and simulate 
        self.m = apiObject
        self.m.switchtolayout()
        self.m.setactivesolver("FDE")
        self.wgWidth = width*1e-6
        self.totalHeight = height*1e-6
        self.etchDepth = etch*1e-6
        self.TE = isTE
        self.λ0 = wavl
        self.Ntrialmodes = trialModes
        self.teThreshold = 0.96
    
    def setSimRegion(self):
        self.m.switchtolayout()
        self.m.setnamed("FDE", "y min", -1.0e-6)
        self.m.setnamed("FDE", "y max", self.totalHeight + 1.0e-6)
        self.m.setnamed("FDE", "x", 0)
        self.m.setnamed("FDE", "x span", self.wgWidth + 2*1.5e-6*self.λ0)
        self.m.setnamed("mesh", "x", 0)
        self.m.setnamed("mesh", "x span", self.wgWidth + 2*1.5e-6*self.λ0)  
            # PMLs general rule of thumb, ~1.5λ from edges of the waveguide core.

    def setBoundaryConds(self):
        """
            PML:      https://support.lumerical.com/hc/en-us/articles/
                            360034382674-PML-boundary-conditions-in-FDTD-and-MODE#toc_2 
            Symmetry: https://support.lumerical.com/hc/en-us/articles/
                            360034382694-Symmetric-and-anti-symmetric-BCs-in-FDTD-and-MODE
        """
        self.m.setnamed("x min bc", "Anti-Symmetric")
        self.m.setnamed("x max bc", "PML")
        self.m.setnamed("y min bc", "Metal")
        self.m.setnamed("y max bc", "Metal")
        self.m.setnamed("same settings on all boundaries", 0)
        self.m.setnamed("pml profile", [1,1])

    def setGeometry(self):
        self.m.switchtolayout()
        self.m.setnamed("structure::guide","x",0)
        self.m.setnamed("structure::pedestal","x",0)
        self.m.setnamed("structure::guide","x span",self.wgWidth)
        self.m.setnamed("structure::guide","y min", 0)
        self.m.setnamed("structure::guide","y max", self.totalHeight)
        self.m.setnamed("structure::pedestal","y min",0)
        self.m.setnamed("structure::pedestal","y max",self.totalHeight-self.etchDepth)
    
    def selectMode(self):
        # Dict of TE polarization fractions corresponding to all trial modes simulated at the FDE
        TEpol = dict([(i,self.m.getdata("mode%i"%i,"TE polarization fraction")) for i in range(1,self.Ntrialmodes)])
        if self.TE==True:
            myTEpol = {key:te for key, te in TEpol.items() if te > self.teThreshold}
        else:
            myTEpol = {key:te for key, te in TEpol.items() if te < 1-self.teThreshold}
        return myTEpol
        
    def solveModes(self):
        self.setGeometry()
        self.setSimRegion()
#         self.setBoundaryConds()
        self.m.run()
        self.Ntrialmodes = 4
        self.m.switchtolayout()
        self.m.set("wavelength", self.λ0*1e-6)
        self.m.setanalysis("number of trial modes", self.Ntrialmodes)
        self.m.findmodes()
        data = self.selectMode()
        return data
    
    def returnWgData(self):
        data = self.solveModes()
        keys = [j for j in data.keys()]
#         Efield = [self.m.getdata("FDE::data::mode%i"%(keys[0]), s)[:,:,0,0] for s in ("Ex","Ey","Ez")]  #E vs x,y,z,lambda
#         Hfield = [self.m.getdata("FDE::data::mode%i"%(keys[0]), s)[:,:,0,0] for s in ("Hx","Hy","Hz")]
#         AeffMode = [self.m.getdata("FDE::data::mode%i"%(keys[0]), "mode effective area")][0]
#         ngMode = [self.m.getdata("FDE::data::mode%i"%(keys[0]), "neff")][0]
#         nMode = [self.m.getdata("FDE::data::mode%i"%(keys[0]), "ng")][0]
        xaxis = self.m.getdata("FDE::data::material", "x")[:,0] * 1e6  
            # second index is because we also have wavl freedom (TODO: allow wavl freedom)
        yaxis = self.m.getdata("FDE::data::material", "y")[:,0] * 1e6
#         nxy = self.m.getdata("FDE::data::material", "index_y")[:,:,0,0]
        #nxy,Efield,Hfield,AeffMode,ngMode,nMode
        return xaxis,yaxis
```

```python
width=3
height=3
etch=1.2

# Open fresh new file, write and simulate 
m = lumapi.MODE("template/waveguide_get_modes.lms",hide=True)
m.setactivesolver("FDE")
wgWidth = width*1e-6
totalHeight = height*1e-6
etchDepth = etch*1e-6
TE = True
λ0 = 2.4

def setSimRegion():
    m.switchtolayout()
    m.setnamed("FDE", "z min", -1.0e-6)
    m.setnamed("FDE", "z max", totalHeight + 1.0e-6)
    m.setnamed("FDE", "y", 0)
    m.setnamed("FDE", "y span", wgWidth + 2e-6)
    m.setnamed("mesh", "y", 0)
    m.setnamed("mesh", "y span", wgWidth + 1e-6)

def setGeometry():
    m.switchtolayout()
    m.setnamed("structure::guide","y",0)
    m.setnamed("structure::pedestal","y",0)
    m.setnamed("structure::guide","y span",wgWidth)
    m.setnamed("structure::guide","z min", 0)
    m.setnamed("structure::guide","z max", totalHeight)
    m.setnamed("structure::pedestal","z min",0)
    m.setnamed("structure::pedestal","z max",totalHeight-etchDepth)

def selectMode():
    # Dict of TE polarization fractions corresponding to all trial modes simulated at the FDE
    TEpol = dict([(i,m.getdata("mode%i"%i,"TE polarization fraction")) for i in range(1,4)])
    if TE==True:
        # Filter out those modes with TE polarization fractions less than 96 percent
        myTEpol = {key:te for key, te in TEpol.items() if te > 0.96}
    else:
        # Filter out those modes with TE polarization fractions less than 6 percent
        myTEpol = {key:te for key, te in TEpol.items() if te < 0.06}
    return myTEpol

def solveModes():
    setGeometry()
    setSimRegion()
    m.run()
    Ntrialmodes = 4
    m.switchtolayout()
    m.set("wavelength", λ0*1e-6)
    m.setanalysis("number of trial modes", 4)
    m.findmodes()
    data = selectMode()
    return data
```

```python
def plotModes():
    data = solveModes()
    keys = [j for j in data.keys()]
    E0_1 = [m.getdata("FDE::data::mode%i"%(keys[0]), s)[0,:,:,0] for s in ("Ex","Ey","Ez")]
    xaxis = m.getdata("FDE::data::material", "y")[:,0] * 1e6  # second index is because we also have wavl freedom.
    yaxis = m.getdata("FDE::data::material", "z")[:,0] * 1e6
    nxy = m.getdata("FDE::data::material", "index_y")[0,:,:,0]

    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.contour(xaxis, yaxis, nxy.T, levels=[3],linewidths=0.5, colors="k")
    im = ax.pcolormesh(xaxis, yaxis, np.real(E0_1[1].T), cmap="jet")
    
    ax.axis([-(width/2)-1,(width/2)+1,-1,4])
    ax.set_xlabel('y [μm]')
    ax.set_ylabel('z [μm]')
    fig.colorbar(im,ax=ax)
    fig.suptitle('Normalized mode distr. $\Re{[E_y]}$', fontsize=15) 
```

```python
plotModes()
```
